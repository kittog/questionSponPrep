```python
import re
from collections import Counter
from typing import List

def nettoyage(texte):
    texte = re.sub(r"[&%!?\|\"{\(\[|_\)\]},\.;/:§»«”“‘…–—−]", "", texte)
    texte = re.sub(r"\d", "", texte)
    texte = texte.replace("’", "'")
    texte = texte.replace("'", "' ")
    return texte

def bag_of_words(text, mots_vides=None, characters=False):
    if mots_vides is None:
        mots_vides = [":",";","!","?","a" ,"à" ,"après" ,"aussi" ,"c'","cet" ,"de" ,"des" ,"en" ,"ensuite" ,"envers" ,"est" ,"et" ,"être" ,"faire"
    ,"font","il","l'","la","le","les","mais","ne","ont","par","pour"]
    text = nettoyage(text)
    if characters:
        return Counter(w for w in text if w and not w.isspace() and w not in mots_vides)
    return Counter(
        l
        for w in filter(None, re.split(r"[\s\.]+", text))
        for l in (w.lower(),)
        if l not in mots_vides
    )

def create_arff_from_questions(questions: List[str], output_file: str):
    mots_vides = [":",";","!","?","a" ,"à" ,"après" ,"aussi" ,"c'","cet" ,"de" ,"des" ,"en" ,"ensuite" ,"envers" ,"est" ,"et" ,"être" ,"faire"
    ,"font","il","l'","la","le","les","mais","ne","ont","par","pour"]

    bows = [bag_of_words(q, mots_vides) for q in questions]
    lexicon = sorted(set(word for bow in bows for word in bow.keys()))
    
    with open(output_file, "w") as out_stream:
        out_stream.write("@relation text_corpus\n")

        for word in lexicon:
            escaped_word = word.replace("'", r"\'")
            out_stream.write(f"@attribute '{{{escaped_word}}}' numeric\n")

        out_stream.write("@data\n")
        for bow in bows:
            out_stream.write(",".join(str(bow.get(word, 0)) for word in lexicon) + "\n")


input_file = 'questions_spontané_clean.txt'  # 请将此处替换为您的文本文件路径
output_file = 'questions_spon.arff'

with open(input_file, 'r') as file:
    questions = [line.strip() for line in file.readlines()]

create_arff_from_questions(questions, output_file)
```
