# mots interrogatifs et d'autres mots

```python
import nltk

with open("questions_spon_clean.txt", 'r') as f:
    sentence = f.read()

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

#annotation POS
tokens = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(tokens)

#cherche des mots interrogatives
question_words = ['qui', 'quoi', 'où', 'quand', 'comment', 'pourquoi', 'est-ce']

question_phrases = []

for i in range(len(pos_tags)):
    if pos_tags[i][0].lower() in question_words:
        question_phrase = " ".join(tokens[i:i+10])  
        question_phrases.append(question_phrase)

print(f"trouve{len(question_phrases)} mot interrogatif：")
for phrase in question_phrases:
    print(f"phrase interrogative：{phrase}")

words = nltk.word_tokenize(sentence)
total_word_count = len(words)
question_word_count = 0

for word in words:
    if word.lower() in question_words:
        question_word_count += 1

print(f"Les pourcentages de mots interrogatifs et d'autres mots sont les suivants：{question_word_count/total_word_count}")
```

# La proportion de phrases contenant des phrases interrogatives dans l'ensemble du texte

```python
import nltk

with open("questions_spon_clean.txt", 'r') as f:
    sentence = f.read()

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

#annotation POS
tokens = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(tokens)

#cherche des mots interrogatives
question_words = ['qui', 'quoi', 'où', 'quand', 'comment', 'pourquoi', 'est-ce']

question_phrases = []

for i in range(len(pos_tags)):
    if pos_tags[i][0].lower() in question_words:
        question_phrase = " ".join(tokens[i:i+10])  
        question_phrases.append(question_phrase)

print(f"trouve{len(question_phrases)} mot interrogatif：")
for phrase in question_phrases:
    print(f"phrase interrogative：{phrase}")

sentences = nltk.sent_tokenize(sentence)
question_sentence_count = 0
total_sentence_count = len(sentences)

for sent in sentences:
    tree = nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(sent)))
    for subtree in tree.subtrees():
        if subtree.label() == 'PERSON':
            print(subtree)
    if any(word.lower() in sent for word in question_words):
        question_sentence_count += 1

print(f"La proportion de phrases contenant des phrases interrogatives dans l'ensemble du texte est de：{question_sentence_count/total_sentence_count}")
```
