# Annotations des textes avec `spaCy`

```python
import spacy
```

```python
import re

def enlever_bruit(question):
    return re.sub(r"#spk[0-9] spk[0-9] : ", "", re.sub(r"#spk[0-9] : ", "", question))
```

```python
for question in toutes_questions:
    print(enlever_bruit(question))
    
```

```python
def morpho_syntaxe(questions):
    nlp = spacy.load("fr_core_news_sm")
    
    # Analyser les dépendances syntaxiques pour chaque question dans le corpus
    docs = [nlp(enlever_bruit(question)) for question in questions]
    
    # Extraire les informations sur les dépendances syntaxiques pour chaque token dans chaque question
    data = [(token.text, token.lemma_, token.pos_) for doc in docs for token in doc]
    
    # Créer un dataframe à partir des informations sur les dépendances syntaxiques
    df = pd.DataFrame(data, columns=["Mot", "lemme", "POS"])
    
    # Regrouper les données en fonction de la colonne "Mot" et compter les occurrences
    df_grouped = df.groupby("Mot").agg({"lemme": "first", "POS": "first"}).reset_index()
    df_grouped["Fréquence"] = df.groupby("Mot").size().reset_index(name="Fréquence")["Fréquence"]
    
    # Trier les données en fonction de la fréquence décroissante
    df_ordered = df_grouped.sort_values(by="Fréquence", ascending=False)
    
    # Supprimer les entrées pour les signes de ponctuation
    df_filtered = df_ordered[(df_ordered["POS"] != "PUNCT")]
    
    return df_filtered
```

```python
df_toutesques = morpho_syntaxe(toutes_questions)
df_toutesques
```

```python
df_annote_spon = morpho_syntaxe(questions_spontanées)
df_annote_spon
```

```python
df_annote_prep = morpho_syntaxe(questions_préparées)
df_annote_prep
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(16, 8))

# Compter le nombre de chaque POS dans chaque dataframe
#pos_counts_all = df_toutesques.groupby("POS").agg({"Fréquence": "sum"}).reset_index()
#total_tokens_all = pos_counts_all["Fréquence"].sum()
#pos_counts_all["Fréquence relative"] = pos_counts_all["Fréquence"] / total_tokens_all

pos_counts_spontaneous = df_annote_spon.groupby("POS").agg({"Fréquence": "sum"}).reset_index()
total_tokens_spontaneous = pos_counts_spontaneous["Fréquence"].sum()
pos_counts_spontaneous["Fréquence relative"] = pos_counts_spontaneous["Fréquence"] / total_tokens_spontaneous

pos_counts_prepared = df_annote_prep.groupby("POS").agg({"Fréquence": "sum"}).reset_index()
total_tokens_prepared = pos_counts_prepared["Fréquence"].sum()
pos_counts_prepared["Fréquence relative"] = pos_counts_prepared["Fréquence"] / total_tokens_prepared

# Tracer un histogramme comparant les fréquences relatives de chaque POS dans les trois dataframes
#plt.bar(pos_counts_all["POS"], pos_counts_all["Fréquence relative"], label="Toutes questions")
plt.bar(pos_counts_spontaneous["POS"], pos_counts_spontaneous["Fréquence relative"], label="questions spontanées")
plt.bar(pos_counts_prepared["POS"], pos_counts_prepared["Fréquence relative"], label="questions préparées")
plt.xlabel("POS")
plt.ylabel("Fréquence relative")
plt.title("Fréquences relatives des différents POS")
plt.legend()
plt.show()
```
